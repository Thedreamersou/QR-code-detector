```python
import numpy as np
import cv2
from pyzbar.pyzbar import decode

# Charger la webcam
cap = cv2.VideoCapture(0)
cap.set(3, 640)  # Largeur
cap.set(4, 480)  # Hauteur

while True:
    success, img = cap.read()
    if not success:
        print("Erreur : Impossible de capturer l'image.")
        break

    # Décodage des QR codes
    for barcode in decode(img):
        print(barcode.data)
        mydata = barcode.data.decode('utf-8')
        print(mydata)
        
        # Dessiner un polygone autour du QR code
        if barcode.polygon:
            pts = np.array([barcode.polygon], np.int32)
            pts = pts.reshape((-1, 1, 2))
            cv2.polylines(img, [pts], True, (255, 0, 255), 5)

            # Afficher le texte décodé
            pts2 = barcode.rect
            cv2.putText(img, mydata, (pts2[0], pts2[1]), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 0.9, (255, 0, 255), 2)

    # Afficher l'image
    cv2.imshow("Result", img)

    # Vérifier si l'utilisateur appuie sur 'q' pour quitter
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Programme terminé.")
        break

# Libérer les ressources
cap.release()
cv2.destroyAllWindows()
```
